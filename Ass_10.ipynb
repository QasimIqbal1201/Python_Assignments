```python
import sqlite3 as sql
import pandas as pd
```

```python
conn = sql.connect('factbook.db')
conn
```

```python
def sql_query(q):
    return pd.read_sql_query(q,conn)
```

```python
q = 'select * from sqlite_master'
sql_query(q)
```

```python
q = 'select * from facts'
sql_query(q)
```

```python
q = 'select * from cities'
sql_query(q)
```

```python
#QUESTION 01 List all cities where the area is greater than 5000, along with their population.
q = '''select c.name as cities, c.population as Population, f.area as Area from cities c
inner join facts f 
on c.facts_id = f.id
where f.area > 5000'''
sql_query(q)
```

```python
# Question 02 List the names of capital cities where the death rate is greater than 2% and the birth rate is less than 20%
q = '''select f.birth_rate as birth_rate, f.death_rate as death_rate, c.capital from facts f
inner join cities c
on f.id = c.facts_id
where c.capital = 1'''
sql_query(q)
```

```python
# Question 03 List the cities with a population growth rate between 1% and 5%, sorted in desc by population growth.
q = '''select c.name as cities, f.population_growth as Population_growth_rate from facts f
inner join cities c
on f.id = c.facts_id
where f.population_growth between 1 and 5
order by f.population_growth DESC'''
sql_query(q)
```

```python
# Question 04 Find countries that start with &quot;A&quot; and do not have any cities listed in the cities table. The query should return the country name and city name. Using left join
q = '''select f.name as Countries, c.name as cities from facts f
left join cities c
on f.id = c.facts_id
where f.name like "A%" and c.name is null
'''
sql_query(q)
```

```python
#Question 05 List all capital cities and their corresponding death rate using a LEFT JOIN.
q = '''select c.name as City, f.death_rate as Death_rate from cities c
left join facts f
on f.id = c.facts_id
where c.capital = 1'''
sql_query(q)
```
