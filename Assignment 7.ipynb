```python
def catagory(experience_coding):
    if experience_coding < 5:
        return "< 5 Years"
    elif experience_coding < 10:
        return "5-10 Years"
    elif experience_coding < 15:
        return "10-15 Years"
    elif experience_coding < 20:
        return "15-20 Years"
    elif experience_coding < 25:
        return "20-25 Years"
    else:
        return "25+ Years"
    
df['Catagory']= df['experience_coding'].apply(catagory)
```
